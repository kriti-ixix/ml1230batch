```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

# Functional Method

```python
plt.plot(x, y, '#EE82EE')
plt.xlabel('My X Array')
plt.ylabel('My Y Array')
plt.title('My Line Graph')
#plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x,y,'g--')
plt.subplot(1, 2, 2)
plt.plot(y,x,'r*-')
```

# Object Oriented Method

```python
#Creating an empty canvas
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y, 'yellow')
```

```python
#Creating a canvas
canvas = plt.figure()

axes1 = canvas.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.set_xlabel("X")
axes1.set_ylabel("Y")
axes1.plot(x, y, 'y')

axes2 = canvas.add_axes([0.6, 0.6, 0.2, 0.2])
axes2.set_xlabel("x")
axes2.set_ylabel("y")
axes2.plot(y, x, 'r')
```
