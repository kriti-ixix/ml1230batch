## Text Preprocessing

```python
import string
```

```python
string.punctuation
```

```python
mess = "Sample message! Hi, how are you? Notice the punctuation."
```

```python
for char in mess:
    print(char)
```

```python
#Filtering out punctuations
nopunc = [char for char in mess if char not in string.punctuation]
```

```python
nopunc
```

```python
nopunc = ''.join(nopunc)
```

```python
nopunc
```

```python
nopunc.split()
```

```python
cleanMessage = [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
```

```python
cleanMessage
```

```python
def textFiltering(mess):
    nopunc = [char for char in mess if char not in string.punctuation]
    nopunc = ''.join(nopunc)
    cleanMessage = [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
    
    return cleanMessage
```

```python
messages.head()
```

```python
messages['message'].head()
```

```python
messages['message'].head().apply(textFiltering)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#Bag of words
bow_transformer = CountVectorizer(analyzer=textFiltering).fit(messages['message'])
```

```python
bow_transformer.vocabulary_
```

```python
len(bow_transformer.vocabulary_)
```

```python
message3 = messages['message'][100]
print(message3)
```

```python
bow3 = bow_transformer.transform([message3])
print(bow3)
```

```python
bow3.shape
```

```python
print(bow_transformer.get_feature_names()[6282])
```

```python
print(bow_transformer.get_feature_names()[10378])
```

```python
messages_bow = bow_transformer.transform(messages['message'])
```

```python
messages_bow.shape
```

```python
messages_bow.nnz
```

```python
# Term Frequency
# number of times the term appears / total terms in the document 

#Inverse Document Frequency
# log_e(total documents / documents with that term in it)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
tfidfTransformer = TfidfTransformer().fit(messages_bow)
```

```python
tfidfTest = tfidfTransformer.transform(bow3)
```

```python
print(tfidfTest)
```

```python
tfidfTransformer.idf_[bow_transformer.vocabulary_['u']]
```

```python
tfidfTransformer.idf_[bow_transformer.vocabulary_['university']]
```

```python
messagesTfidf = tfidfTransformer.transform(messages_bow)
```

```python
messagesTfidf.shape
```

```python
#Naive Bayes Classifier
from sklearn.naive_bayes import MultinomialNB
```

```python
spamDetectModel = MultinomialNB().fit(messagesTfidf, messages['label'])
```

```python
spamDetectModel.predict(tfidfTest)[0]
```
