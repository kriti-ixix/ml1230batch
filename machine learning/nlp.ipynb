## Text Preprocessing

```python
import string
```

```python
string.punctuation
```

```python
mess = "Sample message! Hi, how are you?"
```

```python
for char in mess:
    print(char)
```

```python
#Filtering out punctuations
nopunc = [char for char in mess if char not in string.punctuation]
```

```python
nopunc
```

```python
nopunc = ''.join(nopunc)
```

```python
nopunc
```

```python
nopunc.split()
```

```python
cleanMessage = [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
```

```python
cleanMessage
```

```python
def textFiltering(mess):
    nopunc = [char for char in mess if char not in string.punctuation]
    nopunc = ''.join(nopunc)
    cleanMessage = [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
    
    return cleanMessage
```

```python
messages.head()
```

```python
messages['message'].head()
```

```python
messages['message'].head().apply(textFiltering)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#Bag of words
bow_transformer = CountVectorizer(analyzer=textFiltering).fit(messages['message'])
```

```python
len(bow_transformer.vocabulary_)
```

```python
message3 = messages['message'][3]
print(message3)
```

```python
bow3 = bow_transformer.transform([message3])
print(bow3)
```
